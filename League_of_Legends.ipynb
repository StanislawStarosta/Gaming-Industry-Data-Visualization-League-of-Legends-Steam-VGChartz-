```python
League_age = pd.read_csv("Lol_age.csv", sep=";")
```

```python
length = League_age['Gender'].size
for i in range(length):
    League_age['Percentage of People'].iloc[i] = str(League_age['Percentage of People'].iloc[i])
    League_age['Percentage of People'].iloc[i] = League_age['Percentage of People'].iloc[i].replace('%', '')
    League_age['Percentage of People'].iloc[i] = float(League_age['Percentage of People'].iloc[i])
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

bars = plt.bar(League_age['Gender'], League_age['Percentage of People'], alpha=0.5, edgecolor='black', color=['blue', 'magenta','grey'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')


ax.set_yticks([0, 15, 30, 45, 60, 75, 90])
ax.set_xlabel('Płeć', fontsize=10)
ax.set_ylabel('Dane w [%]', fontsize=10)
plt.title('Podział graczy League of Legends ze względu na płeć', fontsize=12)

plt.savefig('Lol_age_barh.pdf')
plt.show()
```

```python
average_mont_in_year_players = pd.read_csv("amp.csv", sep=";")
```

```python
average_mont_in_year_players
```

```python
average_mont_in_year_players = average_mont_in_year_players.dropna(subset=['Year'])
```

```python
average_mont_in_year_players = average_mont_in_year_players.rename(columns={"Number of average\nmonthly LoL players\nfrom 2011 to 2022":"Number"})
```

```python
for i in range(average_mont_in_year_players['Year'].size):
    average_mont_in_year_players['Year'].iloc[i] = int(average_mont_in_year_players['Year'].iloc[i])
    
    average_mont_in_year_players['Number'].iloc[i] = str(average_mont_in_year_players['Number'].iloc[i])
    average_mont_in_year_players['Number'].iloc[i] = average_mont_in_year_players['Number'].iloc[i].replace(' million', '')
    average_mont_in_year_players['Number'].iloc[i] = float(average_mont_in_year_players['Number'].iloc[i])
```

```python
average_mont_in_year_players = average_mont_in_year_players.sort_values(by='Year')
```

```python
average_mont_in_year_players = average_mont_in_year_players.tail(7)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))


bars = plt.bar(average_mont_in_year_players['Year'], average_mont_in_year_players['Number'], alpha=0.5, edgecolor='black', color="blueviolet")


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')


ax.set_yticks([0, 20, 40, 60, 80, 100, 120, 140, 160])
ax.set_xlabel('rok', fontsize=10)
ax.set_ylabel('Dane w [mln]', fontsize=10)
plt.title('Średnia liczba graczy w miesiącu', fontsize=12)

plt.savefig('Lol_average_numbers_players_in_mont.pdf')
plt.show()
```

```python
revenue = pd.read_csv("LOL_revenue.csv", sep=";")
```

```python
revenue = revenue.rename(columns={"Revenue (In Bn)":"Revenue"})
```

```python
for i in range(revenue['Year'].size):
    revenue['Revenue'].iloc[i] = str(revenue['Revenue'].iloc[i])
    revenue['Revenue'].iloc[i] = revenue['Revenue'].iloc[i].replace('$', '')
    revenue['Revenue'].iloc[i] = float(revenue['Revenue'].iloc[i])
```

```python
fig, ax = plt.subplots(figsize=(8, 5))


bars = plt.bar(revenue['Year'], revenue['Revenue'], alpha=0.5, edgecolor='black', color="darkgoldenrod")


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')


ax.set_yticks(np.arange(0,2.5,0.25))
ax.set_xlabel('rok', fontsize=10)
ax.set_ylabel('Dane w [mld $]', fontsize=10)
plt.title('Zyski League of Legends', fontsize=12)

plt.savefig('Lol_income.pdf')
plt.show()
```
