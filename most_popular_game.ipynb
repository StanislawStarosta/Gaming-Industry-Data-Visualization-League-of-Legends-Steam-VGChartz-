```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
most_popular_games = pd.read_csv("most_popular_games.csv", sep=";")
```

```python
length = most_popular_games['Name'].size
for i in range(length):
    most_popular_games['Number players'].iloc[i] = most_popular_games['Number players'].iloc[i].replace(',', '.')
    most_popular_games['Number players'].iloc[i] = float(most_popular_games['Number players'].iloc[i])
```

```python
most_popular_games = most_popular_games.sort_values(by='Number players', ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(13,8))

bars = ax.barh(most_popular_games['Name'], most_popular_games['Number players'], alpha=0.5, edgecolor='black', color='orange')

for bars in ax.containers:
    ax.bar_label(bars, label_type='edge',color='black', fmt=lambda x: f'{x:.0f}', fontsize=15, position=(5, 0))

ax.set_xlim(0,300)
ax.set_xlabel('Największa liczba graczy w (mln)', fontsize=18)
ax.set_ylabel('Nazwa gry', fontsize=18)
ax.ticklabel_format(axis='x', style='plain', scilimits=(0,0))
ax.tick_params(axis='both', which='both', labelsize=14)

plt.title('Najpopularniejsze gry według komputronik games badane z okresu 30 dni', fontsize=19)

plt.savefig('most_popular_games.pdf', bbox_inches='tight')
plt.show()
```
